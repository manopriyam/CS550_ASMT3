```python
# Importing Required Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from google.colab import drive
```

```python
# Mounting Google Drive
drive.mount('/content/drive')
```

```python
# Importing .csv File from Google Drive Path
file_path = '/content/drive/MyDrive/Colab Notebooks/CS550 Machine Learning Lab/CS550 ASMT 4/WineQT.csv'
```

```python
# Loading the .csv file into a DataFrame
data = pd.read_csv(file_path)
```

```python
# Displaying the Original DataFrame
data.head(10)
```

```python
# Converting the 'quality' Attribute to Binary
data['quality'] = np.where(data['quality'] > 6, 1, 0)
```

```python
# Displaying the Modified DataFrame
data.head(10)
```

```python
# Columns to Normalize
columns_to_normalize = [col for col in data.columns if col not in ['quality', 'Id']]
```

```python
# Initializing the MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Normalizing the Specified Columns
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])
```

```python
# Displaying the Modified DataFrame
data.head(10)
```

```python
# Saving the Modified DataFrame to a .csv file
transformed_file_path = '/content/drive/MyDrive/Colab Notebooks/CS550 Machine Learning Lab/CS550 ASMT 4/WineQT_Modified.csv'
data.to_csv(transformed_file_path, index=False)
```
